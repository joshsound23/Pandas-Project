```python
#Setup
import pandas as pd
from pathlib import Path

load_school_data = Path("Resources/schools_complete.csv")
load_student_data = Path("Resources/students_complete.csv")

#Read into Data Frames
school_data_df = pd.read_csv(load_school_data)
student_data_df = pd.read_csv(load_student_data)

```

```python
school_data_df.head()
```

```python
student_data_df.head()
```

```python
#Merge both data sets
school_student_df = pd.merge(school_data_df, student_data_df, how="left", on=["school_name", "school_name"])
school_student_df.head()
```

```python
#Total unique school
unique_schools=school_student_df['school_name'].nunique()
unique_schools
```

```python
#Total Students
total_students=school_student_df['student_name'].count()
total_students
```

```python
#total_budget
overall_budget=school_student_df.groupby(['school_name'])['budget'].mean()
combined_budget = overall_budget.sum()
combined_budget
```

```python
#Math Average
math_average = school_student_df['math_score'].mean()
math_average
```

```python
#Reading Average
reading_average = school_student_df['reading_score'].mean()
reading_average
```

```python
#math score passing
#math_passing = (school_student_df['math_score']>70).sum() / school_student_df['math_score']
#math_passing_rate = (math_passing / total_students)*100
#math_passing_rate

math_passing=(school_student_df["math_score"]>70).sum() / school_student_df["math_score"].count()* 100
math_passing 
```

```python
#reading score passing
#reading_passing = school_student_df['reading_score'].apply(lambda x: 1 if x >=70 else 0).sum()
#reading_passing_rate = (reading_passing / total_students)*100
#reading_average
reading_pass=(school_student_df["reading_score"]>70).sum() / school_student_df["reading_score"].count()* 100
reading_pass 
```

```python
#DataFrame
district_summary = pd.DataFrame({"Schools": [unique_schools], "Total Students": [total_students], "Total Budget": [combined_budget],
                                "Average Math Score": [math_average], "Average Reading Score": [reading_average],
                                "Math % Passing": [math_passing], "Reading % Passing": [reading_pass], })
district_summary
```

Scores by School Spending

```python
school_spend_sum = school_summary.copy()
school_spend_sum.head()
```

```python
# Establish the bins
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spend_sum["Spending Ranges (Per Student)"] = pd.cut(school_spend_sum['Student Budget'], spending_bins, labels=labels)
school_spend_sum.head()
```

```python
#  Calculate averages for the desired columns. 
spending_math_scores = school_spend_sum.groupby(["Spending Ranges (Per Student)"])["Math Average"].mean()
spending_reading_scores = school_spend_sum.groupby(["Spending Ranges (Per Student)"])["Reading Average"].mean()
spending_passing_math = school_spend_sum.groupby(["Spending Ranges (Per Student)"])["Math Passing"].mean()
spending_passing_reading = school_spend_sum.groupby(["Spending Ranges (Per Student)"])["Reading Passing"].mean()
overall_passing_spending = school_spend_sum.groupby(["Spending Ranges (Per Student)"])["Overall Passing"].mean()
```

```python
# Assemble into DataFrame
spending_summary = pd.DataFrame({"Math Average Score": spending_math_scores, "Reading Average Score":
                                 spending_reading_scores, "passing math average": spending_passing_math,
                                 "Passing Reading Average": spending_passing_reading, "Overall Passing Average": 
                                 overall_passing_spending})

# Display results
spending_summary
```

Scores by School Size

```python
per_school_summary = school_summary.copy()
per_school_summary.head()
```

```python
# Establish the bins.
size_bins = [0, 1000, 2000, 5000]
labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
# Categorize the spending based on the bins
# Use `pd.cut` on the "Total Students" column of the `per_school_summary` DataFrame.

per_school_summary["School Size"] = pd.cut(per_school_summary['Total Students'], size_bins, labels=labels)
per_school_summary.head()
```

```python
# Group the per_school_summary DataFrame by "School Type" and average the results.
average_math_score_by_type = per_school_summary.groupby(["School Type"])["Math Average"].mean()
average_reading_score_by_type = per_school_summary.groupby(["School Type"])["Reading Average"].mean()
average_percent_passing_math_by_type = per_school_summary.groupby(["School Type"])["Math Passing"].mean()
average_percent_passing_reading_by_type = per_school_summary.groupby(["School Type"])["Reading Passing"].mean()
average_percent_overall_passing_by_type = per_school_summary.groupby(["School Type"])["Overall Passing"].mean()
```

```python
# Assemble the new data by type into a DataFrame called `type_summary`
type_summary = pd.DataFrame({"Math Average": average_math_score_by_type, "Reading Average":
                             average_reading_score_by_type, "Passing Math Average": average_percent_passing_math_by_type,
                             "Passing Reading Average": average_percent_passing_reading_by_type, "Overall Passing Average":
                             average_percent_overall_passing_by_type})

# Display results
type_summary
```
